Scrapping the Regitered criminal cases instituted in 2023.

```python
# Importing basic libraries
import numpy as np
import pandas as pd

# Importing a driver from selenium that will help us in scrapping the website
from selenium import webdriver

# Importing service to specify the driver executable path
from selenium.webdriver.chrome.service import Service

# Importing library to locate web element on webpage
from selenium.webdriver.common.by import By

# Importing WebDriverWait for waiting until certain conditions are met
from selenium.webdriver.support.ui import WebDriverWait

# Importing expected_conditions to specify the conditions for WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Importing StaleElementReferenceException to handle stale element reference errors
from selenium.common.exceptions import StaleElementReferenceException


```

```python
# The website we want to scrape
web_to_scrap = "https://njdg.ecourts.gov.in/scnjdg/"

# The path to the Chrome web driver executable
path_to_web_driver = r"G:\STUDY MATERIAL\THEORY\Web scrapping\Scrapping Upwork\chromedriver.exe"
```

```python
# Create a service to specify the driver executable path
service = Service(path_to_web_driver)

# Create ChromeOptions and specify option
chrome_options = webdriver.ChromeOptions()

# Initialize the Chrome WebDriver with service and options
driver = webdriver.Chrome(service=service, options=chrome_options)

# Open the webpage
driver.get(web_to_scrap)


# Locate and select the data related to Pending cases.
pending = driver.find_element(By.XPATH,"//ul[@role='tablist']/li[2]").click()

```

```python
# Wait for up to 10 seconds for the element with ID "loadMe" to become invisible
wait = WebDriverWait(driver, 10)

# Continue once the element with ID "loadMe" is no longer visible
wait.until(EC.invisibility_of_element_located((By.ID, "loadMe")))

# Locate and select the data related to registered cases.
registered = driver.find_element(By.XPATH, "//table[@id='pend_regchart']//td[1]").click()
```

```python
# Locate and select the registered criminal cases 
criminal = driver.find_element(By.XPATH,"//table[@id='OAAE1chart']//td[2]").click()
```

```python
# Locate the registered criminal cases instituted in 2023

inst_vs_Disp = driver.find_element(By.XPATH, "//div[@id='dv_inst_disp_select']//button[@id='insDispTblbtn']")
driver.execute_script("arguments[0].click();", inst_vs_Disp)
```

```python
# Accessing the table that holds information about all the cases

int_disp_table = driver.find_element(By.XPATH, "//table[@id='tblInstDispCount']//tbody//tr[6]//td[2]//a")
driver.execute_script("arguments[0].click();", int_disp_table)
```

```python
# Changing the default number of rows from 10 to 1000

rows_count = driver.find_element(By.XPATH,"//div[@id='tbl_cases_dash_wrapper']//select[@name='tbl_cases_dash_length']//option[@value='1000']").click()
```

```python
# Empty lists to store the data

S_No = []
Diary_Number = []
Case_Number = []
Party_Name = []
```

```python
# Calculating total rows from the in dataset from information available at the bottom of the table

table_info = driver.find_element(By.XPATH, "//div[@id='tbl_cases_dash_info']").text.replace(",", "")
# Filtering and obtaining the total number of entries
total_entries = [i for i in list(table_info.split(" ")) if i.isnumeric()]
total_entries = max(map(int, total_entries))

# Calculating the number of pages based on 1000 entries per page
number_of_pages = (total_entries // 1000) + 1 if total_entries % 1000 != 0 else (total_entries // 1000)



# Looping through each page to extract data

for j in range(number_of_pages):
    # Finding the table on the current page
    table = driver.find_element(By.XPATH, "//table//tbody[@id='caselist_body']")
    
    # Extracting information from each row in the table
    rows = table.find_elements(By.TAG_NAME, "tr")
    for match in rows:
        # Appending data to respective lists
        S_No.append(match.find_element(By.XPATH, "./td[1]").text)
        Diary_Number.append(match.find_element(By.XPATH, "./td[2]").text)
        Case_Number.append(match.find_element(By.XPATH, "./td[3]").text)
        
        # Extracting Party name from a hidden element
        hidden_element = match.find_element(By.XPATH, "./td[4]")
        Party_Name.append(hidden_element.get_attribute("textContent"))

    # Clicking the next button if there are more pages
    if j < number_of_pages - 1:
        next_button = driver.find_element(By.XPATH, "//div[@id='tbl_cases_dash_paginate']//a[3]").click()
        
        # Re-finding the table after clicking the next button
        table = driver.find_element(By.XPATH, "//table//tbody[@id='caselist_body']")
```

```python
# Refresh the webpage to extract information from different table
driver.refresh()
```

Verifying and extracting the other information about each case from other Webpage

```python
# Empty list to store the data

date_of_filing = []
status = []
category =[]
judgement = []
```

```python
import time

# The URL of the website 
web_to_scrap = "https://main.sci.gov.in/case-status"

# The path to the ChromeDriver 
path_to_web_driver = r"G:\STUDY MATERIAL\THEORY\Web scrapping\Scrapping Upwork\chromedriver.exe"

# Service to specify the ChromeDriver executable path
service = Service(path_to_web_driver)

# ChromeOptions to specify additional options for the Chrome browser
chrome_options = webdriver.ChromeOptions()

# Initialize the Chrome WebDriver with the service and options
driver = webdriver.Chrome(service=service, options=chrome_options)

# Open the webpage
driver.get(web_to_scrap)

# Loop through a range of case numbers for scraping
for i in range(4927, 5000):
    # Pause for 2.5 seconds to avoid overloading the website
    time.sleep(2.5)

    # Extract the captcha text from the webpage
    captcha = driver.find_element(By.XPATH, "//p[@id='cap']//font").text

    # Enter the captcha in the input field
    captcha_input = driver.find_element(By.ID, "ansCaptcha").send_keys(captcha)

    # Extract the diary number from your dataframe and enter it on the webpage
    diary_input = driver.find_element(By.XPATH, "//input[@id='CaseDiaryNumber']").send_keys(df.loc[i, "Diary Number"].split("/")[0])

    # Click the button to get the case diary
    case_diary = driver.find_element(By.ID, "getCaseDiary").click()

    # Wait for the table to load on the webpage
    table = WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.XPATH, "//div[@id='collapse1']//tbody")))

    # Extract information from the table and append it to respective lists
    date_of_filing.append(table.find_element(By.XPATH, "//tr[1]//td//div[@width='100%']").text)
    status.append(table.find_element(By.XPATH, "//tr[4]//td[2]").text)
    category.append(table.find_element(By.XPATH, "//tr[6]//td[2]").text)
    judgement.append(table.find_element(By.XPATH, "//tr[5]//td[2]").text)

    # Refresh the webpage to proceed to the next case
    driver.refresh()

# Quit the Chrome WebDriver after scraping all the cases
driver.quit()
```

```python
# Creating a DataFrame to store legal case information
data = pd.DataFrame({
    "Date Filing": date_of_filing,  # Storing the date when the case was filed
    "Status": status,  # Recording the current status of the case
    "Category": category,  # Categorizing the case into a specific legal category
    "Judgement": judgement  # Documenting the judgement related to the case
})



# data.to_csv("SC other half details.csv",index=False)
```
